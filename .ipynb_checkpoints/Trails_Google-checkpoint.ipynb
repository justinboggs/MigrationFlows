```python
# create for loop

# create array of counties 
county_list = counties_df["County Name"]
county_list = county_list.dropna()
county_list = county_list.values

# create array of states 
state_list = counties_df["State"]
state_list = state_list.dropna()
state_list = state_list.values

# create empty list of coordinates 
county_lats = []
county_lng = []

# create base index counter for state
state_index_counter = 0

# begin for loop
for county in county_list:
    
    state = state_list[state_index_counter]
    
    target_county = f"{county}, {state}"

    params = {"address": target_county, "key": gkey}

    # Build URL using the Google Maps API
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    # Run request
    response = requests.get(base_url, params=params)

    # Convert to JSON
    county_response = response.json()

    # Extract lat/lng
    lat = county_response["results"][0]["geometry"]["location"]["lat"]
    lng = county_response["results"][0]["geometry"]["location"]["lng"]

    county_lats.append(lat)
    county_lng.append(lng)
    
    state_index_counter = state_index_counter + 1
    
    print(lat)
```
