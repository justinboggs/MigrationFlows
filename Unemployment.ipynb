```python
import pandas as pd
```

```python
#import unemployement csv file
unemployment="Backup/unemployment data 2018 by county .csv"
```

```python
#create dataframe for the umemployment file
unemployment_df = pd.read_csv(unemployment,header=None)
unemployment_df.head(10)
```

```python
#only pick the columns that needed 
df1=unemployment_df[[3, 4,6,7,8,9]]
df1.head(8)
```

```python
#drop all the non value rows for the dataframe
new=df1.drop(df1.index[0])
new1=new.drop(new.index[0])
new2=new1.drop(new1.index[0])
new3=new2.drop(new2.index[0])
new4=new3.drop(new3.index[1])
new4.columns = new4.iloc[0]
new5=new4.drop(new4.index[0])
new6=new5.dropna()
new6
```

```python
#split county and state from one column to two columns
new6[['County','State']]= new6['County Name/State Abbreviation'].str.split(',', expand=True)
new6
```

```python
#reorganized the order of the columns and get the final data frame
unemployment_df = new6[["County","State","Year","Force","Employed","Unemployed","(%)"]]
unemployment_df
```

```python
#save the csv file
unemployment_df.to_csv('output_data/unemployment.csv', index=False)
```
